## Контроль качества - Проверка диапазонов и типов данных

```python
flight_delays_df['FL_DATE'] = pd.to_datetime(flight_delays_df['FL_DATE'])
print("Столбец 'FL_DATE' успешно преобразован в тип datetime.")
print(f"Тип данных FL_DATE после преобразования: {flight_delays_df['FL_DATE'].dtype}")
```

```python
numeric_cols_to_check = [
    'DEP_DELAY', 'ARR_DELAY', 'TAXI_OUT', 'TAXI_IN', 'AIR_TIME', 'ELAPSED_TIME', 'DISTANCE',
    'CRS_DEP_TIME', 'DEP_TIME', 'CRS_ARR_TIME', 'ARR_TIME'
]

anomalies = {}

for col in numeric_cols_to_check:
    # Define checks based on column nature
    if col in ['DEP_DELAY', 'ARR_DELAY', 'TAXI_OUT', 'TAXI_IN', 'AIR_TIME', 'ELAPSED_TIME', 'DISTANCE']:
        # These values should generally be non-negative. Delays can be negative (early arrival/departure).
        # However, for simplicity, let's assume they shouldn't be excessively negative or large.
        # A typical flight delay/time shouldn't be negative for 'TAXI_OUT', 'TAXI_IN', 'AIR_TIME', 'ELAPSED_TIME', 'DISTANCE'.
        # For DEP_DELAY and ARR_DELAY, negative values mean early, which is valid. We'll check for extremely large or illogical values.
        if col in ['TAXI_OUT', 'TAXI_IN', 'AIR_TIME', 'ELAPSED_TIME', 'DISTANCE']:
            # These should be non-negative. Let's flag anything < 0 as anomaly.
            out_of_range = flight_delays_df[col][flight_delays_df[col] < 0].dropna().shape[0]
        elif col in ['DEP_DELAY', 'ARR_DELAY']:
            # Delays can be negative (early). Check for extreme values, e.g., > 1440 minutes (24 hours) or < -360 minutes (6 hours early)
            # This is a heuristic, actual valid ranges might differ.
            out_of_range = flight_delays_df[col][(flight_delays_df[col] > 1440) | (flight_delays_df[col] < -360)].dropna().shape[0]
        else:
            out_of_range = 0 # Default if no specific check
    elif col in ['CRS_DEP_TIME', 'DEP_TIME', 'CRS_ARR_TIME', 'ARR_TIME']:
        # Time in HHMM format. HH is 00-23, MM is 00-59.
        # So values should be between 0 and 2359 (inclusive).
        # Also, the minute part (last two digits) should not exceed 59.
        # This check applies to both planned and actual times.
        # First, ensure it's not negative or above 2359
        condition1 = (flight_delays_df[col] < 0) | (flight_delays_df[col] > 2359)
        # Second, check if minutes part is > 59
        condition2 = (flight_delays_df[col] % 100) > 59
        out_of_range = flight_delays_df[col][condition1 | condition2].dropna().shape[0]
    else:
        out_of_range = 0 # No specific check for other columns

    if out_of_range > 0:
        percentage = (out_of_range / len(flight_delays_df)) * 100
        anomalies[col] = {'Количество аномалий': out_of_range, 'Процент аномалий': percentage}

if anomalies:
    print("### Обнаруженные аномалии в диапазонах значений по переменным:")
    for col, info in anomalies.items():
        print(f"- {col}: {info['Количество аномалий']} аномалий ({info['Процент аномалий']:.4f}% от общего числа строк)")
else:
    print("Аномалий в диапазонах числовых переменных не обнаружено.")
```

## Разложение временного ряда на тренд и сезонность

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Применяем сезонную декомпозицию
# Учитываем, что данные представлены ежедневно, и недельная сезонность наиболее вероятна (7 дней).
# Используем аддитивную модель, так как амплитуда сезонных колебаний не сильно меняется со временем.
decomposition = seasonal_decompose(daily_flights['Количество рейсов'], model='additive', period=7)

# Строим графики компонентов декомпозиции
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(decomposition.observed)
plt.title('Наблюдаемый ряд')
plt.ylabel('Количество рейсов')

plt.subplot(4, 1, 2)
plt.plot(decomposition.trend)
plt.title('Тренд')
plt.ylabel('Количество рейсов')

plt.subplot(4, 1, 4)
plt.plot(decomposition.resid)
plt.title('Остатки')
plt.xlabel('Дата')
plt.ylabel('Количество рейсов')

plt.tight_layout()
plt.show()
```

## Автокорреляционная функция (ACF)

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

# Построение графика ACF
plt.figure(figsize=(12, 6))
plot_acf(daily_flights['Количество рейсов'], lags=50, ax=plt.gca())
plt.title('Автокорреляционная функция (ACF) для количества рейсов')
plt.xlabel('Лаг')
plt.ylabel('Коэффициент автокорреляции')
plt.grid(True)
plt.show()
```
